### Introduction

In the following notebook, I will be aggregating Inside Aribnb's listings and review data from the last year(05/2019 - 05/2020) into a single file

**Read in libraries**

```python
import pandas as pd
import numpy as np
import glob
```

**Set notebook preferences**

```python
#Increase number of cols displayed
pd.set_option('display.max_columns', 200)
```

**Aggregate and consolidate Listings files**

```python
#Set path to raw data on local machine
path = r'C:\Users\kishe\Documents\Data Science\Projects\Python Projects\In Progress\Airbnb - San Francisco Listings Analysis\Data\01_Raw\SF Airbnb\Raw_Listings_Data'

#Concatenate files in path
listings = pd.concat(map(pd.read_csv, glob.glob(path + '/**')))
```

```python
#Sort columns alphabetically
listings = listings.reindex(sorted(listings.columns), axis=1)

#Drop any duplicate rows
listings.drop_duplicates(inplace = True)

#Check
display(listings.head())
```

**Write file to CSV**

```python
#Print shape of dfs
print('Final shape of aggregated listings data:', listings.shape)

#Set path to write data
path = r'C:\Users\kishe\Documents\Data Science\Projects\Python Projects\In Progress\Airbnb - San Francisco Listings Analysis\Data\01_Raw\SF Airbnb'

#Write aggregated data
listings.to_csv(path + '/2020_0519_Aggregated_Listings.csv')
```
